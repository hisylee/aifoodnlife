```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN

from sklearn.datasets import make_blobs
```

```python
x_data, y_data = make_blobs(n_samples=500, centers = 5, shuffle=True, random_state=10)

print (x_data.shape, y_data.shape)
```

```python
print (x_data [0:10])

print (y_data [0:10])
```

```python
df_cluster = np.unique(y_data)

print (df_cluster)
```

```python
for c in df_cluster: 
    plt.scatter(x_data[y_data == c, 0], x_data[y_data==c, 1], label ='Class{}'.format(c+1))
plt.legend()
plt.show()
```

# DBSCAN

```python
dbscan = DBSCAN(eps=1.1,min_samples=5, metric='euclidean') 

dbscan_pred = dbscan.fit_predict(x_data)
```

```python
print (np.unique(y_data))

print (np.unique(dbscan_pred))
```

```python
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.title('Actual')
for c in np.unique(y_data): 
    plt.scatter(x_data[y_data == c, 0], x_data[y_data==c, 1], label ='Class{}'.format(c+1))
plt.legend()

plt.subplot(1,2,2)
plt.title('Predicted')
for c in np.unique(dbscan_pred): 
    plt.scatter(x_data[dbscan_pred == c, 0], x_data[dbscan_pred==c, 1], label ='Cluster{}'.format(c+1))
plt.legend()
plt.show()
```
