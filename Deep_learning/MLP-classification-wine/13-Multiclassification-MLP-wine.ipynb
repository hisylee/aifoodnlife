```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
```

```python
# 1. Load the data
df = pd.read_excel("wine_dataset.xlsx")

# 2. Separate input data (X) and output labels (y)
x_data = df.iloc[:, :-1].values.astype(np.float32)
y_data = df.iloc[:, -1].values.astype(np.float32)

print (df.shape)
print (x_data.shape)
print (y_data.shape)
```

```python
# 3. Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)

# 4. Normalization
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

print (x_train.shape)
print (x_test.shape)
```

```python
# 5. Build the MLP model

model = Sequential()
model.add(Input(shape = (13,), name = 'input'))
model.add(Dense(32, activation = 'relu', name = 'hidden_1')) 
model.add(Dense(3, activation = 'softmax', name = 'output')) 
```

```python
# 6. Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 7. Train the model
history = model.fit(x_train, y_train, epochs=300, batch_size=16, verbose=1)
```

```python
# 8. Evaluate the model on the test data
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
# 9. Visualize the training process
plt.plot(history.history['loss'], label='Loss', marker='o', linestyle='None')
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.plot(history.history['accuracy'], label='Accuracy', marker='o', linestyle='None')
plt.title('Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
