```python
file = open("abstract-eng.txt", mode='r', encoding='utf-8') # utf-8 or cp949 
doc = file.read()

doc
```

```python
# Word preprocessing

from nltk.tokenize import RegexpTokenizer, WordPunctTokenizer

tokenizer1 = WordPunctTokenizer()

data1 = RegexpTokenizer(r'[a-zA-Z]{3,}').tokenize(doc) # only alphabets, > 3 characters

data2 = [word.lower() for word in data1] # lowercase

print (data2) 
```

```python
# Removal of stop words

from nltk.corpus import stopwords

stopwords = stopwords.words('english')

words = []
for word in data2:
    if word not in stopwords:
        words.append(word)

#print (stopwords)      
print (words)
```

```python
# Removal of stop words and word count

stopwords = ['based', 'high', 'however', 'among', 'utilized', 'end', 'therefore']

words_1 = [word for word in words if word not in stopwords]

word_count = {} 

for noun in words_1 :
    word_count[noun] = word_count.get(noun, 0) + 1

word_count
```

```python
# top_n
from collections import Counter

counter = Counter(word_count)

top20 = counter.most_common(20)
print (top20)
```

```python
# Generation of word clouds
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, background_color='white', max_font_size=250) 
gen = wc.generate_from_frequencies(dict(top20))

plt.figure()
plt.imshow(gen)
```
